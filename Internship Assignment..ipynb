```python
# Importing essential libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
# importing library to get url
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# url of the site
url = "http://stats.espncricinfo.com/ci/content/records/83548.html"
html=urlopen(url)
```

```python
# Using BeautifulSoup to break the html text into python object
soup=BeautifulSoup(html,'lxml')
type(soup)
```

```python
# to get the title
title = soup.title
title
```

```python
# to get the text
text = soup.get_text()
text
```

```python
# To get the rows only
rows=soup.find_all('tr')
print(rows[:89])
```

```python
for row in rows:
    row_td=row.find_all('td')
print(row_td)
```

```python
str_cells=str(row_td)
cleantext=BeautifulSoup(str_cells,"lxml").get_text()
cleantext
```

```python
# "Regular Expression"
import re
list_rows=[]
for row in rows:
    cells=row.find_all('td')
    str_cells=str(cells)
    clean=re.compile('<.*?>')
    clean2=(re.sub(clean,'',str_cells))
    list_rows.append(clean2)
print(clean2)
```

```python
# Now converting list into DataFrame 
df=pd.DataFrame(list_rows)
df.head(89)
```

```python
# Splitting the 0 column into multiple columns using strip method
df1=df[0].str.split(',',expand=True)
df1.head(89)
```

```python
# Stripping the brackets
df1[0] = df1[0].str.strip('[')
df1.head(89)
```

```python
# To get the table headers.
col_labels=soup.find_all('th')
```

```python
# Using BeautifulSoup to get only the headers and not the HTML tags.
all_header=[]
col_str=str(col_labels)
cleantext2=BeautifulSoup(col_str,"lxml").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
# Converting the list of headers into DataFrame.
df2=pd.DataFrame(all_header)
df2.head()
```

```python
# Splitting them like done previously using the Strip Method.
df3=df2[0].str.split(',',expand=True)
df3.head()
```

```python
# Concatenating the two DataFrames.
frames=[df3,df1]
# using concat Method.
df4=pd.concat(frames)
df4.head(89)
```

```python
# Assigning first row to be the header.
df5=df4.rename(columns=df4.iloc[0])
df5.head(89)
```
